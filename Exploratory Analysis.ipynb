```python
# display geo points on the map
center = [44.6252978589571, -90.34580993652344]
zoom = 4
m = Map(center=center, zoom=zoom)

geocircles = []
locations = geo.loc[:,['LATITUDE', 'LONGITUDE']];
for i in range(locations.shape[0]):
    pos = [locations.iloc[i]['LATITUDE'], locations.iloc[i]['LONGITUDE']]
    c = Circle(location=pos, radius=1000)
    geocircles.append(c)
    m.add_layer(c)
   
m
```
